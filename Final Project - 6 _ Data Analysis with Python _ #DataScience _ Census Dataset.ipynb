```python
# calculate the percentage of male workers with secondary school education between the age group 0-29,  
```

```python
# Filter the data for male workers, secondary education, and age group 0-29
male_secondary_education_0_29 = data[(data['Male_Workers'] > 0) & (data['Secondary_Education'] > 0) & (data['Age_Group_0_29'] > 0)]
```

```python
#Calculate the percentage of male workers with secondary education between the age group 0-29
total_male_workers_0_29 = male_secondary_education_0_29['Male_Workers'].sum()
percentage_male_secondary_education_0_29 = (male_secondary_education_0_29['Secondary_Education'].sum() / total_male_workers_0_29) * 100
```

```python
percentage_male_secondary_education_0_29
```

```python
import matplotlib.pyplot as plt

# Create a pie chart
labels = ['Male Workers with Secondary Education (0-29)', 'Other']
sizes = [percentage_male_secondary_education_0_29, 100 - percentage_male_secondary_education_0_29]
colors = ['lightblue', 'lightgray']

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=colors, startangle=90)
plt.title('Percentage of Male Workers with Secondary Education (Age 0-29)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Display the pie chart
plt.show()
```
